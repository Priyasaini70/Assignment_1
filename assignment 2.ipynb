```python
#Question 7:
#create a tuple: data=(5,10,15,20,25,30,35)
# write a program to: count how many numbers are divisible by 5,find sum of all element,find avg of tuple
data = (5, 10, 15, 20, 25, 30, 35)

count = 0
for i in data:
    if i % 5 == 0:
        count += 1
print("element Division by 5:",count)

total = sum(data)
print("Sum of all element:",total)

average = total/len(data)
print("Average of all the element:",average)

```

```python
# Question 8: create a dictonary
# students = {
#    "Aman": 78,
#    "Riya": 92,
#    "Kriti": 88,
#   "Rahul": 95
#}
#Write a program to:Find the student with the highest marks,Find the student with the lowest marks,Print only the students who scored more than 85 

students = {
    "Aman": 78,
    "Riya": 92,
    "Kriti": 88,
    "Rahul": 95
}

highest = max(students, key=students.get)
lowest = min(students, key=students.get)

print("Highest Marks:", highest, students[highest])
print("Lowest Marks:", lowest, students[lowest])

print("Students scoring more than 85:")
for name, marks in students.items():
    if marks > 85:
        print(name, "=", marks)
```

```python
#Question 9:write a function:
#count_frequency(arr) that that takes a list as input and prints the frequency of each element.

def count_frequency(arr):
    freq = {}

    for i in arr:
        if i in freq:
            freq[i] =freq[i]+ 1
        else:
            freq[i] = 1

    print("\n frequency of each element:")
    for key, value in freq.items():
        print(key, "=", value, "times")

n = int(input("E5nter the number of elements: "))

arr = []
for i in range(n):
    element = int(input("Enter a element: "))
    arr.append(element)

count_frequency(arr)
```

```python
 #Question 10:write a function:
#find_duplicates(arr): that takes a list as input and prints all duplicate elements present in the list.
def find_duplicates(arr):
    print("Duplicate elements are:")

    for i in arr:
        if arr.count(i) > 1:
            print(i)
            arr.remove(i)   # prevents printing the same duplicate again

n = int(input("enter number of elements: "))

arr = []
for i in range(n):
    num = int(input("Enter element: "))
    arr.append(num)

find_duplicates(arr)
```
